#Q1. Data Loading and Overview

Load the CSV file into a pandas DataFrame Display the first 10 rows Show the shape of the dataset (rows, columns) Display the column names and their data types

```python
import pandas as pd
students = pd.DataFrame(pd.read_csv("student_data.csv"))
print("Top 10 rows are as follws \n", students.head(10))
print("\n____________________________________________________________________________\n")
print("Dataframe shape is as follws \n", students.shape)
print("\n____________________________________________________________________________\n")
print("Dataframe data types are as follws \n",students.dtypes)
```

#Q2. Summary Statistics

Calculate and display summary statistics for all numerical columns Find the mean, median, and standard deviation of CGPA What is the range (min to max) of AttendancePercent?

```python
#summary
print("Summary of csv fileas follows \n\n",students.describe())
print("\n____________________________________________________________________________\n")

#Find the mean, median, and standard deviation of CGPA
print("Mean of all the students CGPA is : ", f"{students['CGPA'].mean():.2f}")
print("\n____________________________________________________________________________\n")
print("Median of all the students CGPA is : ", f"{students['CGPA'].median():.2f}")
print("\n____________________________________________________________________________\n")
print("Standard deviation of all the students CGPA is : ", f"{students['CGPA'].std():.2f}")
print("\n____________________________________________________________________________\n")

#What is the range (min to max) of AttendancePercent?
print("Min attendance percentage is :", students['AttendancePercent'].min())
print("\n____________________________________________________________________________\n")
print("Max attendance percentage is :", students['AttendancePercent'].max())
```
