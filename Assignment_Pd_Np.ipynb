#Q1. Data Loading and Overview

Load the CSV file into a pandas DataFrame Display the first 10 rows Show the shape of the dataset (rows, columns) Display the column names and their data types

```python
import pandas as pd
students = pd.DataFrame(pd.read_csv("student_data.csv"))
print("Top 10 rows are as follws \n", students.head(10))
def next_line():
  print("\n____________________________________________________________________________\n")
next_line()
print("Dataframe shape is as follws \n", students.shape)
next_line()
print("Dataframe data types are as follws \n",students.dtypes)
```

#Q2. Summary Statistics

Calculate and display summary statistics for all numerical columns Find the mean, median, and standard deviation of CGPA What is the range (min to max) of AttendancePercent?

```python
#summary
print("Summary of csv fileas follows \n\n",students.describe())
next_line()

#Find the mean, median, and standard deviation of CGPA
print("Mean of all the students CGPA is : ", f"{students['CGPA'].mean():.2f}")
next_line()

print("Median of all the students CGPA is : ", f"{students['CGPA'].median():.2f}")
next_line()
print("Standard deviation of all the students CGPA is : ", f"{students['CGPA'].std():.2f}")
next_line()

#What is the range (min to max) of AttendancePercent?
print("Min attendance percentage is :", students['AttendancePercent'].min())
next_line()
print("Max attendance percentage is :", students['AttendancePercent'].max())
```

Q3. Missing Data Check

Check if there are any missing values in the dataset Count the total number of missing values per column Calculate the percentage of missing data for each column

```python
#Check for mission values
print("Checking any missing values in table : \n",students.isnull().any())
next_line()

#Missing values per columns
print("Checking any missing values in column : \n",students.isnull().sum())
next_line()

#Calculate the percentage of missing data for each column
print("Checking percentage of missing values in table : \n",students.isnull().sum() /len(students)*100)
```

#Filtering and Selection (Questions 4-6)

#Q4. Conditional Filtering

1. Find all students with CGPA greater than 3.5 How many students have
2. AttendancePercent below 75%? Display students from the CS department with
3. FinalScore above 80

```python
#students with CGPA greter than 3.5
print("Students with more than 3.5 CGPA : \n",students[students['CGPA']>3.5])
next_line()

#How many students have AttendancePercent below 75%?
print("Students with less than 75% attendance : ",students[students['AttendancePercent']<75].shape[0])
next_line()

#Final score above 80
print("Students with final score above 80 : \n",students[students['FinalScore']>80])
```

#Q5. Data Selection

1. Select only the columns: Name, Department, CGPA, FinalScore
2. Extract rows where students are in semester 5 or above
3. Get the details of students aged 22 or older from the Math department

```python
#Select only the columns: Name, Department, CGPA, FinalScore
print("Printing only limited columns :\n",students[['Name','Department','CGPA','FinalScore']])
next_line()

#Extract rows where students are in semester 5 or above
print("Students from sem 5 and above :\n",students[students['Semester']>5])

#Get the details of students aged 22 or older from the Math department
print("students aged 22 or older from the Math department :\n", students[(students['Age']>=22)&(students['Department']=="Math")])
```

#Q6. Top Performers

1. Find the top 10 students based on CGPA
2. Display the bottom 5 students based on FinalScore
3. Show students who scored above 90 in both MidtermScore and FinalScore

```python
#Find the top 10 students based on CGPA
print("Top 10 students based on CGPA : \n", students.sort_values(by = 'CGPA',ascending=False).head(10))
next_line()

#Display the bottom 5 students based on FinalScore
print("Bottom 5 students based on FinalScore : \n", students.sort_values(by = 'FinalScore',ascending = False).tail(5))
next_line()

#Show students who scored above 90 in both MidtermScore and FinalScore
print("Students who scored above 90 in both MidtermScore and FinalScore : \n", students[(students['MidtermScore']>90)&(students["FinalScore"]>90)] )
```

#Aggregation and Grouping (Questions 7-9)

#Q7. Department-wise Analysis

1. Group data by Department and calculate the average CGPA for each
2. Find which department has the highest average FinalScore
3. Count the number of students in each department

```python
#Group data by Department and calculate the average CGPA for each
print("Department and the average CGPA :\n", students.groupby('Department')['CGPA'].mean())
next_line()

#Find which department has the highest average FinalScore
print("Department with highest average FinalScore :", students.groupby('Department')['FinalScore'].mean().idxmax())
next_line()

#Count the number of students in each department
print("Number of students in each department :\n", students['Department'].value_counts())
```

#Q8. Gender-based Statistics

1. Calculate the average AssignmentScore for each Gender
2. Find the department with the most Female students
3. Compare the mean AttendancePercent across different Genders

```python
#Calculate the average AssignmentScore for each Gender
print("Average AssignmentScore for each Gender :\n", students.groupby('Gender')['AssignmentScore'].mean())
next_line()

#Find the department with the most Female students
print("Department with the most Female students :", students[students['Gender']=='Female']['Department'].value_counts().idxmax())
next_line()

#Compare the mean AttendancePercent across different Genders
print("Mean AttendancePercent across different Genders :\n", students.groupby('Gender')['AttendancePercent'].mean())
```
